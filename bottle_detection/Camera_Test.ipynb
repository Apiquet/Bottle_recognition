```python
import cv2 as cv
import time
import datetime
IMAGES_PATH = "images/"
utils = "Utils/"

bottle_cascade = cv.CascadeClassifier(utils + 'biggest_haar.xml')
cap = cv.VideoCapture('videos/recording_20190605_200212.avi')
```

```python
import cv2
import os

image_folder = 'videos/img'
video_name = 'videos/video_from_imgs.avi'

images = [img for img in os.listdir(image_folder) if img.endswith(".bmp")]
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

video = cv2.VideoWriter(video_name, 0, 1, (width,height))

for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

cv2.destroyAllWindows()
video.release()
```

```python
i = 0
while(cap.isOpened()):
    i = i + 1
    print(i)
    ret, img = cap.read()
    bottles = bottle_cascade.detectMultiScale(img, minNeighbors=6)
    for (x,y,w,h) in bottles:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    time.sleep(1)
    string = "videos/img/frame_" + str(datetime.datetime.now().strftime("%Y%m%d_%H%M%S")) + ".bmp"
    cv.imwrite(string,img)
```
